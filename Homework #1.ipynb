### Question 6

```python
rr = df[df['Make'] == 'Rolls-Royce']
rr
```

```python
rr = rr[
    ['Engine HP', 'Engine Cylinders', 'highway MPG']
  ]
rr
```

```python
rr = rr.drop_duplicates()
rr
```

```python
X = rr.values
X
```

```python
XTX = X.T.dot(X)
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
XTX_inv.sum()
```
